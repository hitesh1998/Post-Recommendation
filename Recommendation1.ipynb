```
import numpy as np
import pandas as pd
post_df=pd.read_csv("posts.csv")
view_df=pd.read_csv("views.csv")
post_df=post_df.rename(columns={"_id": "post_id"})
```

```
post_df=post_df.dropna()
post_df.head()
```

```
view_df.head()
```

```
post_df['category'] = post_df.category.str.split('|')
post_df.head()
```

***For every row in the dataframe, iterate through the list of category and place a 1 into the corresponding column***

```
postWithcategory_df = post_df.copy()
for index, row in post_df.iterrows():
    for ct in row['category']:
        postWithcategory_df.at[index, ct] = 1
#Filling in the NaN values with 0 to show that a post doesn't have that column's category
postWithcategory_df = postWithcategory_df.fillna(0)
postWithcategory_df.head()
```

```
view_df=view_df.drop('timestamp',1)
view_df.head()
```

***Selecting  post id of a perticular user***

```
post=[]
j=0
for i in view_df['user_id']:
    if(i=='5ecb979eeaff6b0c3a58a4f0'):
        post.append(view_df['post_id'][j])
        j+=1
    else:
        j+=1
        
```

```
post
```

geting the DF for the post id that viewed by the user

```
userpost = postWithcategory_df[postWithcategory_df['post_id'].isin(post)]
userpost.head()
```

***Now weight metrics***

```
#dropin the column

userpost = userpost.reset_index(drop=True)
usertable=userpost.drop('post_id',1).drop('title',1).drop('category',1)
usertable=usertable.iloc[:,1:]
usertable
```

```
userProfile = usertable.transpose()
userProfile
```

***sum of weight***

```
userProfile=userProfile.sum(axis=1)
userProfile
```

```
categoryTable = postWithcategory_df.set_index(postWithcategory_df['post_id'])
#And drop the unnecessary information
categoryTable = categoryTable.drop('post_id', 1).drop('title', 1).drop('category', 1)
categoryTable.head()
```

***weight metrics of the all data set***

```
categoryTable=categoryTable.iloc[:,1:]
categoryTable.head()
```

***find the weight of the post ids by user post category***

```
recommendationTable_df = ((categoryTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```

```
#Sort our recommendations in descending order
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
#Just a peek at the values
recommendationTable_df.head()
```
